```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
nltk.download('punkt')


# define the text and word pairs
text = "example text with example word pairs for example word cloud"
word_pairs = [("dbms", "management systems"), ("sql", "primary key"), ("sql", "functions")]

# count the frequency of each word pair
freq_table = {"_".join(pair): text.count(pair[0] + " " + pair[1]) 
              for pair in word_pairs}

# create a heatmap using the frequency table
heatmap_data = pd.DataFrame.from_dict(freq_table, orient='index')
heatmap_data.columns = ['count']
heatmap_data['database'], heatmap_data['features'] = zip(*heatmap_data.index.str.split('_'))
#heatmap_data = heatmap_data.pivot_table(index="database", columns="features", values="count", fill_value=0)
heatmap_data = heatmap_data.pivot_table(index="database", columns="features", values="count", aggfunc='sum', fill_value=0)
sns.heatmap(heatmap_data, cmap="YlGnBu")

# show the plot
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

word_pairs = [("dbms", "management systems"), ("key", "primary key"), ("key", "foreign key"),]

G = nx.Graph()
# add edges to the graph
for pair in word_pairs:
    G.add_edge(pair[0], pair[1])

# set the node positions in a circular layout
pos = nx.circular_layout(G)

# draw the graph
nx.draw(G, with_labels=True, pos=pos)

# show the plot
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# create a graph
G = nx.Graph()

# add nodes to the graph
G.add_node("dbms")
G.add_node("schema")
G.add_node("data mining")
G.add_node("sql")
G.add_node("functions")

# add edges to the graph
G.add_edge("dbms", "schema")
G.add_edge("data mining", "sql")
G.add_edge("sql", "functions")

# set node positions
pos = {
    "dbms": (0, 0),
    "schema": (1, 0),
    "data mining": (1, 1),
    "sql": (2, 0),
    "functions": (2, 1)
}

# set node labels
labels = {
    "dbms": "DBMS",
    "schema": "Schema",
    "data mining": "Data Mining",
    "sql": "SQL",
    "functions": "Functions"
}

# draw the graph with labels and positions
nx.draw(G, pos=pos, with_labels=True, labels=labels)

# show the plot
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt

text = ["sql is a domain-specific language used for managing relational databases",
        "database schema is the structure of a database",
         "dbms allows users to create, manage, and manipulate databases"]

# create a co-occurrence matrix
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(text)
cooc_matrix = (X.T * X).toarray()

# plot the matrix
plt.imshow(cooc_matrix, cmap='hot', interpolation='nearest')
plt.show()
```
